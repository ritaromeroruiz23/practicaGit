## Sin Modulos

````python
shortlist = [3,5,6,2,1,7,9,3,5,7,5]
````

````python
#calculamos la media

def calculate_mean(numbers):
  s=sum(numbers)
  N=len(numbers)
  #calculamos la media
  mean = s/N
  return mean
````

````python
calculate_mean(shortlist)
````

````python
# Calculamos la mediana

def calculate_median(numbers):
    N = len(numbers)
    numbers.sort()

    # encontrar la mediana
    if N % 2 == 0:
      m1 = N/2
      m2 = (N/2) + 1
    # convertir a integer, match position
      m1 = int(m1) - 1
      m2 = int(m2) - 1
      median = (numbers[m1] + numbers[m2])/2
    else:
      m = (N+1)/2
    # convertir a entero, centrar la posicion
      m = int(m) - 1
      median = numbers[m]

    print(" mediana :" + str(median))
    return median
````

````python
print(calculate_median(shortlist))
````

# Combinacion de dataframes

````python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id_new':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
````

````python
# Concat , utilizando filas

pd.concat([left,right], ignore_index= True)
````

````python
# unir por las columnas

pd.concat([left,right], axis=1)
````

````python
# Merge

#Funciona como un JOIN en SQL, combinando datos de dos dataframes en base a una clave.

left.merge(right, on='subject_id')
````

````python
# join
# similira al merge(), pero en lugar de usar una columna como clave, usa el indice.

left_n = pd.DataFrame({
   'a':[1,2,3,4,5],
   'b': [3,4,6,8,7],
   'c':[9,5,3,7,6]},
    index=['i1', 'i2', 'i3', 'i4', 'i6'])
right_n = pd.DataFrame(
   {'d':[1,2,3,4,5],
   'e': [6,5,7,4,3]},
    index=['i1', 'i2', 'i4', 'i5', 'i9'])
````

````python
left_n.join(right_n, how='right')
````
